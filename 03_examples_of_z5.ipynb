```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
sns.set_context('talk', font_scale=2)
sns.set_style('whitegrid')
%matplotlib inline
```

I thought it would be fun to have a little set of examples of Zelazny's Approach to the 5 chart types. Now I'm not sure if it is worth the effort.

```python
plt.figure(figsize=(12,2))

# form 1: pie
plt.subplot(1, 5, 1)
colors = np.array(sns.color_palette(n_colors=3))
plt.pie([15, 25, 60], colors=colors, startangle=90)
plt.savefig('img/zel5-1.png')

# form 2: bar
plt.subplot(1, 5, 2)
plt.bar([0, 1, 2], [15, 25, 60])
plt.axis('off')
plt.savefig('img/zel5-2.png')

# form 3: hbar
plt.subplot(1, 5, 3)
plt.barh([0, 1, 2, 3], [55, 15, 25, 60], color=colors[1])
plt.axis('off')
plt.savefig('img/zel5-3.png')

# form 4: line
plt.subplot(1, 5, 4)
plt.plot([3,1,4,1,5,9,2,6,5,3,5,8], 's-')
plt.axis('off')
plt.axis(xmin=-.5, ymin=-.5, ymax=9.5)
plt.savefig('img/zel5-4.png')

# form 5: scatter
plt.subplot(1, 5, 5)
np.random.seed(12345)
n=25
x = np.random.normal(size=n)
y = np.random.normal(size=n) + x
plt.plot(x, y, 'o', mec=tuple(colors[2]), mew=2, color='none')
plt.axis('off')
plt.savefig('img/zel5-5.png')
```
